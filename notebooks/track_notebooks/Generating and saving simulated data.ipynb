```python
import numpy as np
import matplotlib.pyplot as plt
from narsil2.tracking.cell_simulator.cell import RodShapedCell, CoccoidCell, BentRodCell, EllipsoidCell
from narsil2.tracking.cell_simulator.cell import RectangularCell, RectangularCellStochastic, RodShapedCellStochastic
from narsil2.tracking.cell_simulator.channel import ChannelState
from narsil2.tracking.cell_simulator.cell import CoccoidCellStochastic
from skimage import io, img_as_ubyte
import multiprocessing as mp
from functools import partial
import fastremap
import pickle

from pathlib import Path

%matplotlib qt
```

```python
DATASAVE_PATH = Path('../../data/tracking_data')
```

```python
DATASAVE_PATH.resolve()
```

```python
def generate_datapoint(datapoint_number):

    ecoli_cell = RodShapedCellStochastic(length=46, width=30, position=[40, 40], division_size=84,
                              elongation_rate=4, img_size=(2048, 80))
    channel = ChannelState([ecoli_cell], img_size=(2048, 80), top_boundary=240, bottom_boundary=1600)
    images, links = channel.get_stack(time_points=100, plot=False)

    images_dir = DATASAVE_PATH / 'images' / str(datapoint_number)
    images_dir.mkdir(parents=True, exist_ok=True)
    links_dir = DATASAVE_PATH / 'links' 
    links_dir.mkdir(parents=True, exist_ok=True)
    links_filename = links_dir / (str(datapoint_number) + '.pickle')
    for i in range(images.shape[0]):
        filename = str(i) + '.tiff'
        img = images[i] % 255
        write_path = images_dir / filename 
        io.imsave(images_dir / filename, img_as_ubyte(img), check_contrast=False, 
                 plugin='tifffile', compress=6)
        #print(write_path)
    with open(links_filename, 'wb') as f:
        pickle.dump(links, f)
    
    print(f"{images_dir} -- Done ")
```

```python
number_datapoints = list(range(1000))
```

```python
number_datapoints
```

```python
pool = mp.Pool(processes=6)
pool.map(partial(generate_datapoint,), number_datapoints)
pool.close()
pool.join()
```
