```python
from narsil2.fish.datasets import singleColourFISHData
from skimage.io import imread
from skimage.measure import label, regionprops
from sklearn.mixture import GaussianMixture
from scipy.stats import norm
```

```python
count = {'E.coli': 0, 'E.cocci': 0}

areas = []
for position in range(101, 162):
    position_str = 'Pos' + str(position)
    segmented_filename = Path(analysis_dir_more_cocci) / position_str / 'segmented_phase' / 'img_000000000.tiff'
    image = imread(segmented_filename)
    props = regionprops(label(image))
    for i, blob_prop in enumerate(props, 1):
        if (blob_prop['area'] > 1463 and blob_prop['centroid'][0] < 1740 and blob_prop['centroid'][0] > 350):
            count['E.coli'] += 1
            areas.append(blob_prop['area'])

        elif (blob_prop['area'] < 1463 and blob_prop['centroid'][0] < 1740 and blob_prop['centroid'][0] > 350) :
            count['E.cocci'] += 1
            areas.append(blob_prop['area'])

    print(f"Position: {position} -- Done")
```

```python
count = {'E.coli': 0, 'E.cocci': 0}

areas = []
for position in list(range(201, 268)) + list(range(101, 168)):
    position_str = 'Pos' + str(position)
    segmented_filename = Path(analysis_dir_more_ecoli) / position_str / 'segmented_phase' / 'img_000000000.tiff'
    image = imread(segmented_filename)
    props = regionprops(label(image))
    for i, blob_prop in enumerate(props, 1):
        if (blob_prop['area'] > 1090 and blob_prop['centroid'][0] < 1690 and blob_prop['centroid'][0] > 350):
            count['E.coli'] += 1
            areas.append(blob_prop['area'])

        elif(blob_prop['area'] < 1090 and blob_prop['area'] > 400 and blob_prop['centroid'][0] < 1690 and blob_prop['centroid'][0] > 350):
            count['E.cocci'] += 1
            areas.append(blob_prop['area'])
    print(f"Position: {position} -- Done")
```

```python
data = np.array(areas)
gmm = GaussianMixture(n_components = 3).fit(data.reshape(-1, 1))
plt.figure()
plt.hist(data, bins=500, histtype='stepfilled', density=True, alpha=0.5)
#plt.xlim(0, 360)
f_axis = data.copy().ravel()
f_axis.sort()
a = []
for weight, mean, covar in zip(gmm.weights_, gmm.means_, gmm.covariances_):
    a.append(weight*norm.pdf(f_axis, mean, np.sqrt(covar)).ravel())
    plt.plot(f_axis, a[-1])
plt.plot(f_axis, np.array(a).sum(axis =0), 'k-')
plt.xlabel('Variable')
plt.ylabel('PDF')
plt.tight_layout()
plt.show()
```
